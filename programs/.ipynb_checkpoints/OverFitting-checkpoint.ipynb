```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import models
%matplotlib inline
```

```python
plt.style.use("grayscale")
plt.style.use("seaborn-whitegrid")
gray = (150/255,150/255,150/255)
```

```python
N = 20
sigma2_true = 0.1

# x = np.random.normal(size=N)
x = np.random.uniform(low=-1, high=1, size=N)
func = lambda x: np.sin(x*3) + x - 0.3*x**2
y = func(x) + np.random.normal(scale=np.sqrt(sigma2_true), size=N)

plt.grid(color=gray)
plt.scatter(x, y, s=2, c="k")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

多項式基底関数を過剰に大きな次元に設定する

```python
polydim = 100
phi = models.PolyBaseFunction(x, dims=polydim)
model = models.LinearRegression(y, phi)
```

```python
model.fit_OLS()
```

```python
xmin, xmax = x.min(), x.max()
xline = np.linspace(xmin, xmax, num=50)
philine = models.PolyBaseFunction(xline, dims=polydim)
```

```python
plt.grid(color=gray)
plt.scatter(x, y, s=2, c="k", label="data")
plt.plot(xline, model.predict_OLS(philine), c="gray", label="prediction")
plt.ylim(y.min()-0.1, y.max()+0.1)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.savefig("../draft/img/OverFitting.png", dpi=300)
plt.show()
```

適切に正則化をするか，ベイズ法を用いれば，滑らかに内挿される

ただし，いずれも追加されるパラメータを適切に調整する必要がある

```python
model.fit_Ridge(alpha=0.001) # 緩めの正則化
model.fit_Bayes(sigma2=0.1, prior_cov=10*np.eye(model.D)) # 緩めの事前分布で，観測誤差の分散は真値
```

```python
plt.grid(color=gray)
plt.scatter(x, y, s=2, c="k")
plt.plot(xline, model.predict_Ridge(philine), label="Ridge", linestyle=":")
plt.plot(xline, model.predict_Bayes(philine)[0], label="Bayes", linestyle="-.")
plt.ylim(y.min()-0.1, y.max()+0.1)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
